```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import itertools
```

```python
label_count = 4
window_size = 3
```

```python
G = nx.DiGraph()

labels = [str(x+1) for x in range(label_count)]

for per in itertools.permutations(labels, window_size):
    ids = "".join(per)
    G.add_node(ids)

for per in itertools.permutations(labels, window_size-1):
    base_ids = "".join(per)
    for start_id in [id for id in labels if id not in per]:
        start_ids = start_id + base_ids
        
        for end_id in [id for id in labels if id not in per]:
            end_ids = base_ids + end_id

            G.add_edge(start_ids, end_ids)
   
```

```python
layout = nx.spring_layout(G)


nx.draw_networkx_edges(G, pos = layout)
nx.draw_networkx_nodes(G, pos = layout)
nx.draw_networkx_labels(G, pos = layout)
```
